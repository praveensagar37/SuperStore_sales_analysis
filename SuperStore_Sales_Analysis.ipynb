## Exploratory Data Analysis

* Univariate analysis

```python
# for categorical columns (data_categorical_column2)
for i in data_categorical_columns2: 
    plt.figure(figsize=(15,7))
    plt.xticks(rotation= 75)
    sns.countplot(x= data[i])
    plt.title(i)
```

```python
sns.heatmap(data.corr(), annot=True)
```

* Bivariate analysis

#### Analysing profit and total sales in different regions

```python
region_wise_1 = data.groupby(['Region'],as_index= False).agg({'Sales':'sum','Profit':'sum'})
region_wise_1
```

```python
plt.figure(figsize=(15,12))

plt.subplot(2,2,1)
sns.barplot(x= region_wise_1['Region'], y=region_wise_1['Sales']);
plt.title("Region vs Total Sales")

plt.subplot(2,2,2)
sns.barplot(x= region_wise_1['Region'], y=region_wise_1['Profit']);
plt.title("Region vs Total Profits")

plt.subplot(2,2,3)
plt.pie(region_wise_1['Sales'], labels=region_wise_1['Region'], autopct='%1.1f%%', startangle=90);
plt.title("Total Sales %")

plt.subplot(2,2,4)
plt.pie(region_wise_1['Profit'], labels=region_wise_1['Region'], autopct='%1.1f%%', startangle=90);
plt.title("Total Profits %")
```

#### Observations:- Total sales is highest in 'West'(33.4%) and lowest in 'South'(16.1%) region. Total profits is higest in 'West'(38.7%) and lowest in 'South'(15.1%) region.

#### Analysing profit and total sales in different category and region wise

```python
region_wise = data.groupby(['Region','Category'],as_index= False).agg({'Sales':'sum','Profit':'sum'})
region_wise
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
sns.heatmap(region_wise.pivot(index="Region", columns="Category", values="Sales"), annot= True);
plt.title("TOTAL SALES")

plt.subplot(1,2,2)
sns.heatmap(region_wise.pivot(index="Region", columns="Category", values="Profit"), annot= True);
plt.title("TOTAL PROFIT")
```

#### Observations:- From TOTAL SALES plot, it is clear that 'SOUTH' region has lowest as sales compared to other region. From 'TOTAL PROFIT' plot we can see that 'Furniture' category is least profitable.

```python
plt.figure(figsize=(10,6))
sns.barplot(x= region_wise['Region'], y= region_wise['Sales'], hue= region_wise['Category'])
plt.title("Region vs Sales")
```

#### Observations:- Products in 'Office Supplies' category has higest sales in all regions.

```python
plt.figure(figsize=(10,6))
sns.barplot(x= region_wise['Region'], y= region_wise['Profit'], hue= region_wise['Category'])
plt.title("Region vs Profit")
```

#### Observations:- 'Furniture' category in Central region is in Loss and least profitable compared to other category products.

### Analysing sales and profit with respect to Category column

```python
category_wise = data.groupby('Category', as_index= False).agg({'Sales':'sum','Quantity':'sum','Returns':'mean','Profit':'sum'})
category_wise
```

```python
plt.figure(figsize=(15,12))

plt.subplot(2,2,1)
sns.barplot(x= category_wise['Category'], y= category_wise['Sales'])
plt.title("category vs sales")

plt.subplot(2,2,2)
sns.barplot(x= category_wise['Category'], y= category_wise['Profit'])
plt.title("category vs profit")

plt.subplot(2,2,3)
plt.pie(category_wise['Quantity'], labels=category_wise['Category'], autopct='%1.1f%%', startangle=90);
plt.title("category vs Quantity")

plt.subplot(2,2,4)
sns.barplot(x= data['Category'], y= data['Returns'])
plt.title("category vs Returns")
```

#### Observations:- Quantity_wise, 'Office Supplies' is the most sold product category (61.1%) and least sold is 'Technology'(18.2%) and 'Technology' is most profitable category. 'Furniture category falls in most returned items.

### Analysing Sub-Categories with respect to Quantity, Sales and Profits

```python
plt.figure(figsize=(15,19))

plt.subplot(3,1,1)
sub_category = data.groupby(['Category','Sub-Category'], as_index= False)[['Quantity']].sum()
plt.xticks(rotation= 75)
sns.barplot(x= sub_category['Sub-Category'], y= sub_category['Quantity'], hue= sub_category['Category'])

plt.subplot(3,1,2)
sub_cat_sales = data.groupby(['Category','Sub-Category'], as_index= False)[['Sales']].sum()
plt.xticks(rotation= 75)
sns.barplot(x= sub_cat_sales['Sub-Category'], y= sub_cat_sales['Sales'], hue= sub_cat_sales['Category'])

plt.subplot(3,1,3)
sub_cat_profits = data.groupby(['Category','Sub-Category'], as_index= False)[['Profit']].sum()
plt.xticks(rotation= 75)
sns.barplot(x= sub_cat_profits['Sub-Category'], y= sub_cat_profits['Profit'], hue= sub_cat_profits['Category'])
```

#### Observations:- In Sub-category 'Tables' has negative impact on profit.

### Analysing Sales and Profit with respect to Segment

```python
segment = data.groupby('Segment', as_index= False).agg({'Sales':'sum','Quantity':'sum','Returns':'mean','Profit':'sum'})
segment
```

```python
plt.figure(figsize=(15,6))

plt.subplot(1,2,1)
plt.pie(segment['Sales'], labels=segment['Segment'], autopct='%1.1f%%', startangle=90);
plt.title("Sales vs Segment")

plt.subplot(1,2,2)
plt.pie(segment['Profit'], labels=segment['Segment'], autopct='%1.1f%%', startangle=90);
plt.title("Profit vs Segment")
```

#### Observations:- 48.1% of total sales comes from 'consumer segment', 32.6% of total sales comes from 'Corporate' and 19.4% of total sales comes from 'Home Office'. Highest profit i.e. 46.4% comes from 'consumer' segment and lowest profit i.e. 20.6% is generated by 'home office' segment.

### Analysing sales and profit in different states

```python
state_data = data.groupby(['Region','State'], as_index= False).agg({'Sales':'sum','Quantity':'sum','Returns':'mean','Profit':'sum'})
state_data
```

```python
plt.figure(figsize=(15,12))

plt.subplot(1,2,1)
sns.barplot(x= state_data['Sales'], y= state_data['State'], hue= state_data['Region'])
plt.title("TOTAL SALES")

plt.subplot(1,2,2)
sns.barplot(x= state_data['Profit'], y= state_data['State'], hue= state_data['Region'])
plt.title("TOTAL PROFITS")
```

#### observations:- From above plot we can infer that CALIFORNIA has highest sales and profit in WEST region, NEWYORK has highest sales and profit in EAST region. TEXAS has highest sales in CENTRAL region but posses losses, similarly FLORIDA has highest sales in SOUTH region but has losses.

```python
data_texas=data[data['State']=='Texas'].groupby(['Category']).agg({'Sales':'sum','Quantity':'sum','Returns':'max','Profit':'sum'})
data_texas
```

#### Obsevations:- In Texas profits in 'Furniture' , 'Office Supplies' and 'Technology' categories are Negative and the reason could be High Returns of items.
